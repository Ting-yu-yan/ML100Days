# 文章標題 :100層序列推薦模型也能被加速訓練



文章出處:https://www.leiphone.com/category/academic/EaQYc7IAg2kXCq4Z.html

**專案目標:**

因為目前大部分的推薦系統模型都是採用較淺層的神經網絡，一來是資料的數量龐大 (交易紀錄可能高達數百意)
，二來是層數越深越容易產生 gradient vanish or gradient explosure 的問題，因此我們往往都採用較淺
層的神經網絡，然而淺層的神經網絡往往代表著預測準確度不佳，因此本論文的目的就是提出一個能用在推薦系
統的深層網絡模型。

**使用的技術:**

在 res block 中增加一個參數調整 residual 的留存量，作者發現該方法有效地使得訓練的過程更加穩定，意即較不易發生梯度爆炸或是消失的問題，二來作者發現，一個完整的神經網絡模型中，中間的 res block 彼此之間的相關性很高，這點說明這些 block 貌似可以共同使用一組參數，因此大大地降低參數的訓量數量以及參數的存取空間。 

具體方法: 

1）預訓練一個淺層序列推薦模型；
2）對該模型進行層堆疊，得到一個兩倍深度的模型；
3）微調這個深層模型；
4）將深層模型作為一個新的淺層模型，重複1）至3）直到滿足業務需求。
