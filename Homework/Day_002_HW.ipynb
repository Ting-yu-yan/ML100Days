## [作業目標]
了解機器學習適合應用的領域與範疇。

## [作業重點]
透過瞭解機器學習的運作方式，找出非常適用機器學習應用的領域及其原因

## 作業
請找出一個你認為機器學習可以發光發熱的領域並闡述原因? 

在過去，受限於硬軟體的限制，我們沒有辦法利用幾十萬甚至千萬的資料來訓練模型，因此機器學習的方法大多都是從數學模型的假設
、分佈的假設、變數之間的關係假設以及變數的限制開始，當我們的實際問題和我們的模型假設越符合，我們越能透過少量的資料來分
析並處理問題，然而當問題越發複雜，我們開始引入深度學習的模型，來處理問題，舉凡人臉辨識、車牌辨識、自動化妝、自動更換顏
色等等，然而深度學習跟傳統上機器學習最大的不同在於深度學習需要訓練的參數可能高達數百萬計，因此必須仰賴龐大的訓練資料，
同時模型的結構也相對非常複雜，我們難以直觀地解讀。因此我認為，機器學習可以發光發熱的領域其實非常多種，尤其是當我們需要
在小樣本下處理問題時，機器學習的許多方法可以說是比深度學習的模型還要更實際且有用，舉凡統計品管、實驗設計、訊號處理與分類
等我認為都是屬於機器學習的一環，且這些學門在實際上的應用可多了，舉凡透過腦波分析患者的睡眠品質，透過實驗設計找出重要的
機台參數、透過統計品管偵測出機台產生變異的時間點，以及透過迴歸分析或離散分析來建立分類模型，找出潛在消費者，甚至還能找出
關鍵的影響變數。綜合以上我認為，機器學習與深度學習在本質上有著非常大的不同，其中最重要的差別在於問題的複雜度以及模型的複雜度，
越是複雜的問題搭配上複雜的模型，我們越是需要大量的資料以及深度學習的技巧，相反地，越是簡單的問題以及簡單的模型，我們可以
透過少量的資料配合機器學習的技巧來處理。
